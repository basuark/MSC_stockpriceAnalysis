```python
def fibonacci_counter(datadf):
    mydict ={}
    levels = ['level_0','level_38','level_50','level_61','level_78','level_100']
    for i in levels:
        mydict[i]=0
    
    for i in range(len(datadf)-1):
        if datadf['Trend'][i]==-1 and datadf['Trend'][i+1]==1:
            for j in levels:
                if datadf['Low'][i]>=datadf[j][i]-0.005 :
                    mydict[j]+=1
        elif datadf['Trend'][i]==1 and datadf['Trend'][i+1]==-1:
            for j in levels:
                if datadf['High'][i]<=datadf[j][i]+0.005 :
                    mydict[j]+=1
    return mydict   
                
```

```python
def simulation(datadf):
    mydict ={}
    levels = ['level_0','level_38','level_50','level_61','level_78','level_100']
    for i in levels:
        mydict[i]=0
    curr_amt=10000
    final_amt=0
    total_invested=0
    total_return=0
    for i in range(len(datadf)-1-7):
        totalmtbuy=0
        totalmtsell=0
        if datadf['Trend'][i]==-1 and datadf['Trend'][i+1]==1:
            if np.mean(datadf['FibonacciLevels'][i+1:i+7]) > datadf['FibonacciLevels'][i] :
                buyprice=datadf['FibonacciLevels'][i]
                quantity=curr_amt//datadf['FibonacciLevels'][i]
                totalmtbuy += buyprice*quantity
                
        elif datadf['Trend'][i]==1 and datadf['Trend'][i+1]==-1:
            if np.mean(datadf['FibonacciLevels'][i+1:i+7]) < datadf['FibonacciLevels'][i] :
                sellprice=datadf['FibonacciLevels'][i]
                quantity=curr_amt//datadf['FibonacciLevels'][i]
                totalmtsell += sellprice*quantity
        total_invested+=totalmtbuy
        total_return+=totalmtsell
    return total_invested,total_return
                
```
