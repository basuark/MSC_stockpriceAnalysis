```python
apollohosp_fibonacci = pd.read_csv(DATA_HOME+'\\apollohosp\\apollohosp_fibonacci.csv')
apollohosp_fibonacci.head()
```

```python
apollohosp_master = apollohosp_indicators.merge(apollohosp_fibonacci,left_on='Date',right_on='Date',how='inner', suffixes=('','_fib'))
apollohosp_master.head()
```

```python
def adjust_Targets(datadf):
    length = len(datadf)
    datadf['LeadingClose']=np.nan
    datadf['LeadingFibonacciLevel']=np.nan
    for i in range(length-1):    
        datadf['LeadingClose'][i] = datadf['Close'][i+1]
        datadf['LeadingFibonacciLevel'][i] = datadf['FibonacciLevels'][i+1]
    return datadf
apollohosp_master=adjust_Targets(apollohosp_master)
```

```python
apollohosp_master['Date'] = pd.to_datetime(apollohosp_master['Date'],format='%Y-%m-%d')
apollohosp_master =apollohosp_master.set_index('Date')
```

```python
apollohosp_master=apollohosp_master.loc['2016-01-01':'2021-11-30']
print(apollohosp_master.index.min(),apollohosp_master.index.max())
```

```python
apollohosp_master.to_csv(DATA_HOME+'\\apollohosp\\apollohosp_master_fullbkp.csv')
```

```python
drop_cols=[ 'Open', 'High', 'Low', 'Dividends','Stock Splits',  'PrevYear',
           'Unnamed: 0', 'Open_fib', 'High_fib', 'Low_fib', 'Close_fib', 'Volume_fib', 'Dividends_fib',
           'Stock Splits_fib', 'SwingHigh', 'SwingLow', 'Trend', 'AdjSwingHigh', 
           'AdjSwingLow', 'level_0', 'level_38', 'level_50', 'level_61','level_78', 'level_100', 
           'level_127', 'level_161', 'level_261', 'val','diff']
apollohosp_master = apollohosp_master.drop(drop_cols,axis=1)

apollohosp_master.info()
```

```python

apollohosp_master.to_csv(DATA_HOME+'\\apollohosp\\apollohosp_master.csv')
```
