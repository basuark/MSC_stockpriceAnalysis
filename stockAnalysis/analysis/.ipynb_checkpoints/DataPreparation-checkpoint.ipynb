```python
import numpy as np
import pandas as pd

import warnings
warnings.filterwarnings("ignore")
```

```python
from Scripts import *
print(DATA_HOME)
```

```python
# Merging Financial and Technical Indicator dataframes into single dataframe

APOLLOHOSP_DIR=DATA_HOME+'\\apollohosp'
DABUR_DIR = DATA_HOME + '\\dabur'
MARUTI_DIR = DATA_HOME+'\\maruti'
TCS_DIR = DATA_HOME + '\\tcs'
```

```python
apollohosp_findata = pd.read_csv(APOLLOHOSP_DIR+'\\apollohosp_financial_data.csv')
apollohosp_tech_ind = pd.read_csv(APOLLOHOSP_DIR+'\\apollohosp_technicalindicators.csv')
```

```python
dabur_findata = pd.read_csv(DABUR_DIR+'\\dabur_financial_data.csv')
dabur_tech_ind = pd.read_csv(DABUR_DIR+'\\dabur_technicalindicators.csv')
```

```python
maruti_findata = pd.read_csv(MARUTI_DIR+'\\maruti_financial_data.csv')
maruti_tech_ind = pd.read_csv(MARUTI_DIR+'\\maruti_technicalindicators.csv')
```

```python
tcs_findata = pd.read_csv(TCS_DIR+'\\tcs_financial_data.csv')
tcs_tech_ind = pd.read_csv(TCS_DIR+'\\tcs_technicalindicators.csv')
```

```python
apollohosp = apollohosp_findata.merge(apollohosp_tech_ind, on="Date", how="inner",suffixes=('','_y'))
apollohosp.head()
```

```python
apollohosp = apollohosp.drop(['Unnamed: 0','Unnamed: 0_y',
       'Open_y', 'High_y', 'Low_y', 'Close_y', 'Volume_y', 'Dividends_y',
       'Stock Splits_y'],axis=1)
apollohosp.head()
```

```python
dabur = dabur_findata.merge(dabur_tech_ind, on="Date", how="inner",suffixes=('','_y'))
dabur = dabur.drop(['Unnamed: 0','Unnamed: 0_y',
       'Open_y', 'High_y', 'Low_y', 'Close_y', 'Volume_y', 'Dividends_y',
       'Stock Splits_y'],axis=1)
maruti = maruti_findata.merge(maruti_tech_ind, on="Date", how="inner",suffixes=('','_y'))
maruti = maruti.drop(['Unnamed: 0','Unnamed: 0_y',
       'Open_y', 'High_y', 'Low_y', 'Close_y', 'Volume_y', 'Dividends_y',
       'Stock Splits_y'],axis=1)
tcs = tcs_findata.merge(tcs_tech_ind, on="Date", how="inner",suffixes=('','_y'))
tcs = tcs.drop(['Unnamed: 0','Unnamed: 0_y',
       'Open_y', 'High_y', 'Low_y', 'Close_y', 'Volume_y', 'Dividends_y',
       'Stock Splits_y'],axis=1)
```

```python
apollohosp.to_csv(APOLLOHOSP_DIR+'\\apollohosp_merged.csv',index=False)
dabur.to_csv(DABUR_DIR+'\\dabur_merged.csv',index=False)
maruti.to_csv(MARUTI_DIR+'\\maruti_merged.csv',index=False)
tcs.to_csv(TCS_DIR+'\\tcs_merged.csv',index=False)
```
