```python
# Reading Financial data from data/apollohosp folder
apollohosp_findata = pd.read_csv(APOLLOHOSP_DIR+"\\Financialdata.csv")
apollohosp_findata.head()
```

```python
apollohosp['PrevYear'] = apollohosp['Date'].apply(lambda x: int(x[0:4]))
apollohosp.head()
```

```python
apollohosp_fin_1=apollohosp.merge(apollohosp_finratio,left_on='PrevYear',right_on='Year')
apollohosp_fin_1=apollohosp_fin_1.drop(['PrevYear','Year'],axis=1)
apollohosp_fin_1.head()
```

```python
# Creating a dictionary to map months to quarters
mnthtoquarterdict={
            'Mar':'04',
            'Jun':'01',
            'Sep':'02',
            'Dec':'03'}
```

```python
apollohosp_findata['QuarterYear']=apollohosp_findata['MonthYear'].apply(lambda x: str('20'+x[0:2]))+apollohosp_findata['MonthYear'].apply(lambda x: x[3:6]).map(mnthtoquarterdict)
```

```python
def mnth_to_leadingquarter(mnth):
    quarter=0
    if int(mnth)>=1 and int(mnth)<4:
        quarter='03'
    elif int(mnth)>=4 and int(mnth)<7:
        quarter='04'
    elif int(mnth)>=7 and int(mnth)<10:
        quarter='01'
    elif int(mnth)>=10 and int(mnth)<13:
        quarter='02'
    return quarter
```

```python
apollohosp['QuarterYear'] =  apollohosp['Date'].apply(lambda x: x[0:4])+apollohosp['Date'].apply(lambda x:mnth_to_leadingquarter(x[5:7]))
apollohosp[['Date','QuarterYear']].head()
```

```python
apollohosp_fin_merged = apollohosp.merge(apollohosp_findata,on="QuarterYear",how="inner")
apollohosp_fin_merged = apollohosp_fin_merged.drop(['QuarterYear','MonthYear'],axis=1)
apollohosp_fin_merged.head()
```
