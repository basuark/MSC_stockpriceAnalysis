Data Merger

```python
# Importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import datetime as dt

pd.options.display.max_columns=None
```

```python
# Reading saved data 
apollohosp = pd.read_csv("apollohosp_stockpricedata.csv")
apollohosp.head()
```

```python
# Reading Financial ratios from data/apollohosp folder
apollohosp_finratio = pd.read_csv("data/apollohosp/FinancialRatios.csv")
apollohosp_finratio.head()
```

```python
# Reading Financial data from data/apollohosp folder
apollohosp_findata = pd.read_csv("data/apollohosp/Financialdata.csv")
apollohosp_findata.head()
```

```python
apollohosp.info()
```

```python
apollohosp['Year'] = apollohosp['Date'].apply(lambda x: int(x[:4]))
apollohosp['Month'] = apollohosp['Date'].apply(lambda x: x[5:7])
apollohosp.head()
```

```python
apollohosp_fin_1 = apollohosp.merge(apollohosp_finratio,on="Year",how="inner")
apollohosp_fin_1.head()
```

```python
# 
mnthdict={'Jan':'01',
            'Feb':'02',
            'Mar':'03',
            'Apr':'04',
            'May':'05',
            'Jun':'06',
            'Jul':'07',
            'Aug':'08',
            'Sep':'09',
            'Oct':'10',
            'Nov':'11',
            'Dec':'12'}
apollohosp_findata['Month'] = apollohosp_findata['MonthYear'].apply(lambda x: x[0:3])
apollohosp_findata['Year'] = apollohosp_findata['MonthYear'].apply(lambda x: int('20'+x[4:]))

apollohosp_findata['Month'] = apollohosp_findata['Month'].map(mnthdict)
apollohosp_findata.head()
```

```python
apollohosp_fin = apollohosp_fin_1.merge(apollohosp_findata,on=["Year","Month"],how="inner")
apollohosp_fin.head()
```

```python
apollohosp_fin[['Year','Month','MonthYear']].value_counts()
```

```python
apollohosp_fin.columns
```

```python
apollohosp_fin = apollohosp_fin.drop(['Year','Month','MonthYear'],axis=1)
apollohosp_fin.to_csv('apollohosp_financials.csv',index=False)
```
